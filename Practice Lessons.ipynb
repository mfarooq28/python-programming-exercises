```python
import numpy as np

x1 = np.sort(np.array([1,2,3.1]))
x2 = np.sort(np.array([3,2,1]))

deltas = np.sum((x1-x2)**2)
print(f'Numpy Deltas = {deltas}')

x1 = [1,2,3]
x2 = [3.1, 2, 1]

x1.sort()
x2.sort()
deltas = sum([(x1-x2)**2 for x1, x2 in zip(x1, x2)])

print (f'Our Pyton Deltas = {deltas}')
```
